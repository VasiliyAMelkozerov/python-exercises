# Conditionals

## if this then that

```python
x = 6

r  = x % 2

if r == 0:
    print("Our number is even")
    print("Grettings!")
print('Complete')
```

```python
x = 5

r  = x % 2

if r == 0:
    print("Our number is even")
    if x > 2:
        print('Greater then two')
    else:
        print('less than or equal to 2!')
    
if r == 1:
    print("Our number is odd!")
    
print('Complete')
```

```python
#nested if's

x = 13

r  = x % 2

if r == 0:
    print("Our number is even")
    if x > 2:
        print('Greater then two')
    else:
        print('less than or equal to 2!')
    
if r == 1:
    print("Our number is odd!")
    
print('Complete')
```

```python
x = 6

if x == 1:
    print('ONE')
elif x == 2:
    print('TWO')
elif x == 3:
    print('THREE')
else:
    print('Hurray!')
```

## While

### The while loop  will execute it's body until the condition evaluates to false

```python
i = 1
while i < 10:
    print(i)
    #i=i+i
```

```python
stored_password =  'password123'
user_input = input('Please enter a password: ')

while user_input != stored_password:
    print('Incorrect password')
    user_input = input('Please enter a password:')
    
print("You may now proceed")
```

```python
user_input = 12  
i = 1
available_candies  = 7

while i <= user_input:
    if i > available_candies:
        print('Sorry, out of candies')
        break
    print(i, "Here   is your candy")
    i += 1
```

##  Continue

```python
#print only odd  numbers:
l = [1,2,4,6,2,12,44,15,17]
for x in l:
    if x % 2 == 0:
        continue
    print(x,"here is an odd number")
```
